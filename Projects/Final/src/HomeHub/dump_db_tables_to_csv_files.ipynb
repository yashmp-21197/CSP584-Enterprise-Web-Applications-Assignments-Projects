```python
import contextlib
import csv
import datetime
import os
import pymysql

USERS_SQL_QUERY = """
SELECT * FROM USERS;
"""
CATEGORIES_SQL_QUERY = """
SELECT * FROM CATEGORIES;
"""
SERVICES_SQL_QUERY = """
SELECT * FROM SERVICES;
"""
TRANSACTIONS_SQL_QUERY = """
SELECT * FROM TRANSACTIONS;
"""

connection = pymysql.connect(host='localhost',
                             user='root',
                             password='root',
                             db='home_hub')

with contextlib.closing(connection):
    with connection.cursor() as cursor:
        cursor.execute(USERS_SQL_QUERY)
        users_results = cursor.fetchall()

        cursor.execute(CATEGORIES_SQL_QUERY)
        categories_results = cursor.fetchall()

        cursor.execute(SERVICES_SQL_QUERY)
        services_results = cursor.fetchall()

        cursor.execute(TRANSACTIONS_SQL_QUERY)
        transactions_results = cursor.fetchall()

users_output_file = 'CSVs/users-{}.csv'.format(datetime.datetime.today().strftime('%Y-%m-%d'))
with open(users_output_file, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile, lineterminator='\n')
    csv_writer.writerows(users_results)

categories_output_file = 'CSVs/categories-{}.csv'.format(datetime.datetime.today().strftime('%Y-%m-%d'))
with open(categories_output_file, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile, lineterminator='\n')
    csv_writer.writerows(categories_results)

services_output_file = 'CSVs/services-{}.csv'.format(datetime.datetime.today().strftime('%Y-%m-%d'))
with open(services_output_file, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile, lineterminator='\n')
    csv_writer.writerows(services_results)

transactions_output_file = 'CSVs/transactions-{}.csv'.format(datetime.datetime.today().strftime('%Y-%m-%d'))
with open(transactions_output_file, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile, lineterminator='\n')
    csv_writer.writerows(transactions_results)
```

```python
import csv
import datetime
import os
import pymongo

client = pymongo.MongoClient("mongodb://localhost:27017/")

db = client["HomeHubReviews"]
col = db["myReviews"]

csv_columns = {"_id", "title", "service_id", "user_id", "user_name", "user_age", "user_gender", "review_rating", "review_date", "review_text"}

document = col.find()

myreviews_output_file = 'CSVs/myReviews-{}.csv'.format(datetime.datetime.today().strftime('%Y-%m-%d'))

try:
    with open(myreviews_output_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in document:
            writer.writerow(data)
except IOError:
    print("I/O error")
```
